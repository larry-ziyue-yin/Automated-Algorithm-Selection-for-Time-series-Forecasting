```python
import matplotlib.pyplot as plt
from sklearn.metrics import ConfusionMatrixDisplay, confusion_matrix

mse_true = [0.363, 0.405, 0.425, 0.414, 0.286, 0.352, 0.377, 0.406, 0.281, 0.326, 0.348, 0.399, 0.17, 0.23, 0.282, 0.373]
mae_true = [0.382, 0.408, 0.423, 0.435, 0.34, 0.384, 0.409, 0.427, 0.339, 0.368, 0.399, 0.418, 0.25, 0.291, 0.327, 0.386]

def build_labels(true, predict):
    labels = []
    for i in range(len(predict)):
        if true[i] == predict[i]:
            labels.append(1)
        else:
            labels.append(0)
    return labels

def draw_cm(mse_predict, mae_predict, title):
    mse = build_labels(mse_true, mse_predict)
    mae = build_labels(mae_true, mae_predict)
    cm = confusion_matrix(mse, mae)
    
    disp = ConfusionMatrixDisplay(confusion_matrix=cm)
    disp.plot(cmap='GnBu')
    
    plt.title(title, fontsize=14)
    plt.xlabel('MSE Prediction', fontsize=12)  # 修改X轴标签
    plt.ylabel('MAE Prediction', fontsize=12)
    plt.show()
```
